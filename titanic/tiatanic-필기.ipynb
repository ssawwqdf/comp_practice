<pre>
* apply와 transform 차이  
* groupby 값을 이용해 fillna()할 경우 transform() 사용
* train.groupby(" ")[''].apply()    vs     train.groupby(" ")[''].transform()

[차이1]
apply: 그룹 by 한 거에서도 개별컬럼에 접근 가능
transform: 그룹 by 한 데이터프레임에서는 개별컬럼에 접근 불가능

[차이2]
apply: 연산 안 되는 애 있으면 에러
transform: 알아서 연산 가능한 애만 골라주기는 한데 사실 데이터프레임 전체에 적용하는 경우는 잘 없어서 상관 ㄴㄴ


[차이3] ***
연산 가능한 컬럼만 함수 적용한다면
apply: 그룹별 데이터프레임 출력 -> 이래서 group by 하면 데이터프레임 작아짐 
transform: (멀티인덱스)	기존 인덱스 유지 -> 이래서 group by 해도 원래 인덱스 그대로 유지 / 알아서 값 꺼내서 사이사이에 끼워넣는다

```python
#    package      package (label.py) class         class
from sklearn.preprocessing import LabelEncoder, OneHotEncoder

data_list=['b','a','c','d','a','b'] # 순서 뒤죽박죽해도 알파벳순서대로 sorting한다. -> 굳이 내 마음대로 하고 싶으면 dict로 해서 map 써라
lb_encoder=LabelEncoder()
lb_encoder.fit(data_list) #이것부터 해야함에 유의!!
res=lb_encoder.transform(data_list)
print("unique하게 바꿀 대상 선별(fit)한 것: ", lb_encoder.classes_)
print("글자를 숫자로 바꾼(transform) 결과(인코딩 결과): ",res)
print("원상복구(inverse)(숫자를 원래 글자로): ", lb_encoder.inverse_transform(res))

# fit이랑 transform 따로 하기 귀찮은 경우
# lb_encoder=LabelEncoder()
# res=lb_encoder.fit_transform(data_list)

# 근데 왜 따로 만들었을까?? -> 중요한 이유가 있다.(머신러닝 할 때 알려주겠다.)
# train과 test셋이 보통 따로 있다. -> fitting할 대상이 달라지면 train에서 학습했을 때 없던 데이터가 test에 있을 수도 있다.
# 그러면 train에서 학습한대로는 test에 적용할 수 없기 때문에 문제가 생긴다.


# .fit () 어디보자... -> fit할 대상 데이터(array-like) 넣음. -> return self (나한테 return 주는 게 아니라 메모리 쫓아가서 바꿔는 놓고 주소만 주겠다.) . unique 뭐 있나 물색만 함.
# 물색 끝나면 원래 주소 안에 뭐 바꿀지 알고 있으니까 거기다 transform()함.(근데 알고 있는데 굳이 또 다시 값 지정해줘야됨.) -> return array-like로 y값 준다. 글자를 숫자로 바꾼 게 온다. 1차로
# fit뭐 물색했는지 보겠다. classes_
# 결과 보고 싶으면 어디 담아서 res
# 원본으로 돌려놔라가 inverse_transform(결과)

```

```python
def my_score(y_true, y_pred):
    report=classification_report(y_true, y_pred)
    accuracy=accuracy_score(y_true, y_pred)
    precision=precision_score(y_true, y_pred)
    recall=recall_score(y_true, y_pred)
    f1=f1_score(y_true, y_pred)
    
    print("<REPORT>")
    print(report, '\n')
    print('--'*30)
    print("<DETAIL>")
    print(f"\t accuracy:\t{accuracy:.4f} \n \t precision:\t{precision:.4f} \n \t recall:\t{recall:.4f} \n \t f1:\t\t{f1:.4f}")
    print('--'*30)
```
