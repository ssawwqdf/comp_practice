# Processing(mini)
* 컬럼 이름 소문자
* ID 삭제
* Specises unique값 확인

```python
# lower columns_name
lower_col=[col.lower() for col in iris.columns]
lower_col
iris.columns=lower_col
iris.columns
```

```python
# delete ID
iris=iris.drop('id', axis=1)
iris.head()
```

```python
# unique of feature columns
np.unique(iris['species'])
# value_counts로도 가능!!
```

```python
# species -> int
col_dict={'Iris-setosa':1, 'Iris-versicolor':2, 'Iris-virginica':3}
iris['species']=iris['species'].map(col_dict)
iris['species'].value_counts()
```

```python
# columns rename
iris.columns=['sepal_length', 'sepal_width', 'petal_length', 'petal_width','class']
iris.head()
```

```python
# use -> str.explicit??
```

# Check score

```python
X=iris.drop('class', axis=1)
y=iris['class']
```

```python
X_train, X_test, y_train, y_test = train_test_split(X, y, random_state=1024, test_size=0.3) # test_size 조정 가능
```

```python
print(X_train.shape, X_test.shape, y_train.shape, y_test.shape)
```

```python
model=RandomForestClassifier(n_estimators=100, random_state=1024)
model.fit(X_train, y_train)
pred=model.predict(X_test)
f1=f1_score(y_test, pred, average='macro')
acc=accuracy_score(y_test, pred)
print(f"f1: {f1:.4f} accuracy: {acc:.4f}")
cm=confusion_matrix(y_test, pred)
print(cm)
```

# Lightgbm  
https://github.com/microsoft/LightGBM/blob/master/examples/python-guide/sklearn_example.py

```python
import lightgbm as lgb
# lgb.LGBMClassifier

# lgb.쓰기 귀찮으니까 Classifier까지 import
from lightgbm import LGBMClassifier
```

```python
X=iris.drop('class', axis=1)
y=iris['class']
```

```python
X_train, X_test, y_train, y_test = train_test_split(X, y, random_state=1024, test_size=0.3) # test_size 조정 가능
```

```python
# 규제 필요하면 GBoost 쓰고 속도 필요하면 lightgbm 쓰면 된다.

model=LGBMClassifier(max_depth=5, learning_rate=0.1, n_estimators=100, random_state=1024)


model.fit(X_train, y_train) #, eval_set=[(X_test, y_test)] # 점수도 내부에서 설정 가능
pred=model.predict(X_test)

f1=f1_score(y_test, pred, average='macro')
acc=accuracy_score(y_test, pred)

print(f"f1: {f1:.4f} accuracy: {acc:.4f}")

cm=confusion_matrix(y_test, pred)
print(cm)
```
