```python

# titanic에서 불러온 거 그대로 실행하면 에러메세지 Target is multiclass but average='binary'. Please choose another average setting, one of [None, 'micro', 'macro', 'weighted'].
# average='binary'가 기본값 -> average는 [None, 'micro', 'macro', 'weighted'] 중에 하나 골라줘! (accuracy_score만 빼고) -> 설명 노션

def my_score(y_true, y_pred):
    report=classification_report(y_true, y_pred)
    accuracy=accuracy_score(y_true, y_pred)
    precision=precision_score(y_true, y_pred,    average='macro') # 보통 macro로 써서 macro로 수정
    recall=recall_score(y_true, y_pred,          average='macro')
    f1=f1_score(y_true, y_pred,                  average='macro')
    print('--'*30)
    print("<REPORT>")
    print(report, '\n')
    print('--'*30)
    print("<DETAIL>")
    print(f"\t accuracy:\t{accuracy:.4f} \n \t precision:\t{precision:.4f} \n \t recall:\t{recall:.4f} \n \t f1:\t\t{f1:.4f}\n")

def my_fit_score(train, target="credit"):
    model_list = [RandomForestClassifier(n_estimators=30, random_state=1234) , # n_estimators는 너무 느려서 일단 30으로 줄이겠다.
              LogisticRegression(penalty="l2", random_state=1234),
              KNeighborsClassifier(),
              #DecisionTreeClassifier(),
              #LinearSVC()
             ]
    
    y = train[target]
    X = train.drop([target], axis=1)

    X_train,X_test, y_train, y_test = train_test_split(X, y, 
                                                   test_size=0.2, 
                                                   random_state=1234,
                                                   shuffle=True
                                                   )
   
    for model in model_list:
        model.fit(X_train, y_train)  #---학습해라
        pred = model.predict(X_test)
        # acc_score = accuracy_score(y_test, pred)
        print('=='*30)
        print(f"\t\t<<{model.__class__.__name__}>>")
        my_score(y_test, pred)
        
        #  print(model.feature_importances_) # 어레이로 출력됨. => 모델이 어떤 변수를 중요하게 봤나
        # -------------------------피쳐 중요도
        # plt.rc('font', family='Malgun Gothic')
        # plt.xticks(rotation=90) # 90도로 돌림
        # plt.bar(X.columns, model.feature_importances_)
    print('=='*30)

# 
```

### KFold()

```python
# cross-val은 stratified다.
# kfold=KFold(n_splits=5, shuffle=False, random_state=1024)
# KFold는 y 안 줘도 그만(어차피 순서대로 자르니까)

model = RandomForestClassifier(n_estimators=100, random_state=1234)
kfold = KFold(n_splits=5) #, shuffle=False, random_state=1234)
f1_list = []
for train_index, test_index  in kfold.split(X, y) :  #train_index, test_index
    #X_train, X_test,y_train, y_test = train_test_split(X,y, test_size=0.2)
    X_train, X_test = X.iloc[train_index], X.iloc[test_index]
    y_train, y_test = y.iloc[train_index], y.iloc[test_index]
    model.fit(X_train, y_train)
    pred = model.predict(X_test)
    f1 = f1_score(y_test, pred, average="macro")
    f1_list.append(f1)
    
print( f"f1 평균 점수 : {np.array(f1_list).mean():.4f} ")
```

### StratifiedKFold()

```python
model = RandomForestClassifier(n_estimators=100, random_state=1234)
skfold = StratifiedKFold(n_splits=5, shuffle=True, random_state=1234) #----------아까는 KFold()
f1_list = []
for i, (train_index, test_index)  in enumerate(skfold.split(X, y)) :  #train_index, test_index
    #X_train, X_test,y_train, y_test = train_test_split(X,y, test_size=0.2)
    X_train, X_test = X.iloc[train_index], X.iloc[test_index]
    y_train, y_test = y.iloc[train_index], y.iloc[test_index]
    model.fit(X_train, y_train)
    pred = model.predict(X_test)
    f1 = f1_score(y_test, pred, average="macro")
    print(f"{i}Fold : {f1}")
    f1_list.append(f1)
print( f"f1 평균 점수 : {np.array(f1_list).mean():.4f} ")
```

파라미터를 최적화 할 수 있게 도와줌 -> 모델의 점수 잘 나오게 파라미터 조정해주겠다. (학습에 좋은 모델 골라준다!!  내부적으로 Cross_val_score한다. 그래서 cv 필요)  
모델은 RandomForestClassifier, ... 이런 걸 의미 -> 해당 모델이 어떤 파라미터 갖는지 알면 좋다.(근데 파라미터 조정하는 건 겁나 어렵다.)

  
refit : 좋은 파라미터 찾았으면 모델에 적용해줄까? 당근 -> True값이 기본  
cv 그냥 주면 계속 변하니까 skfold에서 주는 걸로 주겠다.  
param_grid는dict꼴로 달라.  
return : gcv object로 준다.  
Attributes: `.cv_results` 하면 dict로 각 방법 별 점수(values는 어레이에 넣어서) 준다.  
각 키를 이용해서 점수 빼낼 수 있다.(API doc 참고)  
함수로 제일 좋은 점수 이런 것도 빼낼 수 있다.  

`class sklearn.model_selection.GridSearchCV(estimator, param_grid, *, scoring=None, refit=True, cv=None)`

```python
# 이건 필기 
# RandomForestClassifier 모델 써서 해보자(문법 연습!!!!!!) -> 파라미터 설명 노션 참고
# 강사님은 모델 먼저 나눴는데 물론 상관은 없겠지만 모델 순서상으로는 gcv한 다음에 들어가야할듯?
# depth 너무 짧게 주면 모델 점수 안 좋을 수도 -> 어설프게 건들 거면 기본값 써라 튜닝하지 말고

# y=train[]

# model=RandomForestClassifier(n_estimators=100, random_state=1234)

# #--- 하이퍼 파라미터: 일반적인 파라미터 말고 모델의 성능에 영향 주는 걸 하이퍼파라미터라고 한다.
# # 하이퍼 파라미터 조정:: 튜팅
# mypram={'n_estimators': [100, 200, 300],
#        'max_depth': [3,4,5],
#         'min_samples_split': [2,3,4]
#        } # 이거 중에 하나 써서 제일 점수 줘

# # 100으로도 한 번 fit 200으로도 한 번 fit 300으로도 한 번 fit한다. 

# gcv=GridSearchCV()
```

```python
y = train['credit']
X = train.drop(['credit'], axis=1)
X_train,X_test, y_train, y_test = train_test_split(X, y, 
                                                   test_size=0.2,
                                                   random_state=1234,
                                                   shuffle=True
                                                      )

model = RandomForestClassifier(n_estimators=100, random_state=1234)
#-- 튜닝 : 하이퍼파라미터 값을 임의로 변경해서 모델의 성능을 좋게 한다.
myparam = {'n_estimators':[100,200],
           'min_samples_split' : [1,2,3,4]
        }

skfold = StratifiedKFold(n_splits=5, shuffle=True, random_state=1234) 
gcv_model = GridSearchCV(model, param_grid=myparam, scoring="f1_macro", refit=True, cv=skfold)


gcv_model.fit(X_train, y_train)
print(gcv_model.best_score_)
print(gcv_model.best_params_)
# print(gcv_model.cv_results_)

```

```python
pred = gcv_model.predict(X_test)
f1 = f1_score(y_test, pred, average='macro')
print(f"GCV f1 점수 {f1:.4f}")   # 원본 f1 0.5278    S.Kfold: 0.5277
```
