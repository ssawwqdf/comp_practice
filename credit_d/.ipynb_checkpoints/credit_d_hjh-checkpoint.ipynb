## label encode

```python
def encs(df):
    enc=LabelEncoder()
    for i in df.columns.tolist():
        if df[i].dtype=='object':
            enc.fit(df[i])
            df[i]=enc.transform(df[i])
    return df
```

```python

encs(train)
```

## 새 train 만들고, 모델 생성

```python
#trains=pd.concat([train2,news],axis=1)
```

```python
train.head()
#train=train.drop(['child_num','DAYS_BIRTH'],axis=1)
train=train.drop('FLAG_MOBIL',axis=1)
```

```python
y = train['credit']
X = train.drop(['credit'], axis=1)
```

```python
X_train,X_test, y_train, y_test = train_test_split(X, y, 
                                                   test_size=0.2, 
                                                   random_state=1234,
                                                   shuffle=True,
                                                   stratify=y
                                                   )
print(X_train.shape, X_test.shape, y_train.shape, y_test.shape)
```

```python
model_list = [RandomForestClassifier(n_estimators=150, random_state=105) ,
              LogisticRegression(penalty="l2", random_state=42),
              KNeighborsClassifier(),
              DecisionTreeClassifier(),
              LinearSVC()
             ]
for model in model_list:
    model.fit(X_train, y_train)  #---학습해라
    내답안지 = model.predict(X_test)
    acc_score = accuracy_score(y_test, 내답안지)
    print(f"{model.__class__.__name__} \t\t 정확도:{acc_score:.4f}")
```
